```c++
#include <iostream>
#include <vector>
#include <algorithm>

// Inline computations (no function definitions)
int a = 1, b = 5, c = 3;
int max_val = std::max(a, std::max(b, c));
std::cout << "max=" << max_val << std::endl;

std::vector<double> v{1.0, 2.0, 3.0};
double s = 0.0;
for (double x : v) s += x;
double mean = v.empty() ? 0.0 : s / v.size();
std::cout << "mean=" << mean << std::endl;
```
