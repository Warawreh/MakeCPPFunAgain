Intro to `try/catch` and `assert` for basic error handling.

```c++17
#include <iostream>
#include <stdexcept>
#include <cassert>

assert(2 + 2 == 4);
try {
  int a = 10;
  int b = 2;
  if (b == 0) throw std::runtime_error("divide by zero");
  std::cout << (a / b) << std::endl;
} catch (const std::exception& e) {
  std::cerr << e.what() << std::endl;
}
```
